# 1. Venta de un Producto:

Diseñar un programa que genere una factura de compra para un producto. El programa debe solicitar
al usuario:

• Nombre del cliente.

• Identificación del cliente.

• Nombre del producto.

• Precio unitario del producto.

• Cantidad comprada.

**La factura debe mostrar:**

• Nombre de la tienda.

• Número de factura "Factura de Venta No. 000"

• Datos del cliente "nombre e identificación"

• Detalles del producto "nombre, precio unitario, cantidad comprada"

• Resumen de la compra "base, IVA del 19% y valor total"

---

Algoritmo

1. inicia
2. declar las variables como corresponda (string, int o real)
3. mostrar mensaje para que el usuario sepa que ingresar
4. leer las variables que el usuario introdujo
5. calcular el valor a pagar multiplicando el precio del producto por la cantidad
6. calcular el valor del iva
7. mostrar la informacion como una factura
8. finalizar

```
program fact

start

  declare string name
  declare string  id

  declare real unit_price

  display "ingrese el nombre del producto"
  input product_name
  display "ingrese el precio unitario"

  input q
  display "ingrese la fecha de compra"
  input date

  set cant_pagar= unit_price *q
  set iva= cant_pagar *0.19

  display "nombre de la empresa"

  display "N_0001"

  display "precio de  compra con impuesto iva", iva
  display "cantidad a pagar sin iva ", cant_pagar

end
```

````
name= str(input("ingrese su nombre: "))
id= str(input("ingresar su identificación: "))
product_name=str(input("ingrese el nombre del producto: "))
unit_price=float(input("ingrese el precio unitario: "))
q=int(input("cantidad de unidades de ese producto: "))
date=str(input("ingrese la fecha de compra: "))

cant_pagar= unit_price * q
iva= cant_pagar*0.19
total= cant_pagar + iva

print("spender nova ")
print("factura electrónica: ")
print("N_0001: ")
print("identificación del comprador: " , id)
print("detalle: " , product_name)
print("cantidad de unidades del producto comprado", q)
print("precio de  compra con impuesto iva", total)
print("cantidad a pagar sin iva ", cant_pagar)
print(date)
````

# 2. Calcular Utilidad:

El programa debe solicitar al usuario:

• Precio de compra del producto.

• Precio de venta del producto.


**El programa debe mostrar:**

• Margen de operación y administrativo 15% del precio de
 venta"

• Costo del producto

- precio de venta + margen de operación y administrativo.

• Utilidad absoluta

-diferencia entre el precio de venta y el costo del producto

• Utilidad porcentual.

---
algoritmo
1. iniciar el programa
2. declarar las variables
3. mostrarle al usuario en mensaje para que el usuario sepa que ingresar
4. leer la infomacion del usuario

8. calcular la utilidad porcentual
9. mostrar los resultados
10. finalizar

```
program util

```

````
p_compra=float(input("ingrese el valor de compra: "))
p_venta=float(input("ingrese el valor de venta del producto: "))

margen =p_venta *0.15
costo= p_compra + margen
u_adsoluta=p_venta-costo
u_porcentual=(u_adsoluta/costo)*100

print("Margen de operacion: " , margen )
print("Costo del producto: " , costo)
print("la utilidad absoluta: ", u_adsoluta )
print("la utilidad porcentual del producto es: ", u_porcentual)
````
